```python
from unidecode import unidecode
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Dataset original

accidentalidad_2021 = pd.read_excel('../data/alcaldia/Accidentalidad-2021-2023.xlsx', sheet_name='2021')
accidentalidad_2021 = accidentalidad_2021.iloc[:, :25].copy()
accidentalidad_2021
```

```python
# Modificaciones
accidentalidad_2021.rename(columns={
    "Unnamed: 3": "Direccion",
    " ": "18-30", "30-60": "31-60",
    "Unnamed: 20": "Vehiculo",
    'D.D': 'Dia', 'M.M': 'Mes',
    'A.A': 'Año'
}, inplace=True)

accidentalidad_2021.columns = list(map(lambda x: unidecode(str(x).strip().capitalize()), accidentalidad_2021.columns))
accidentalidad_2021 = accidentalidad_2021.iloc[:-1]

nan_cols = list(set(accidentalidad_2021.columns) - set(['Dia', 'Mes', 'Ano', 'Direccion', 'Vehiculo']))

for col in nan_cols:
    accidentalidad_2021.loc[:, nan_cols] = accidentalidad_2021[nan_cols].fillna(0)
```

## Estandarizacion por columnas

```python
# Columna "Solo daños"
accidentalidad_2021.loc[accidentalidad_2021['Solo danos'] == ' ', 'Solo danos'] = 0
accidentalidad_2021.loc[:, 'Solo danos'] = accidentalidad_2021['Solo danos'].astype(float)

# Columna "Heridos"
accidentalidad_2021.loc[:, "Herido"] = accidentalidad_2021["Herido"].replace(130.0, 0.0)

# Columna "Muerto"
accidentalidad_2021.loc[:, "Muerto"] = accidentalidad_2021["Muerto"].replace(13.0, 0.0)

# Columna "Dia"
accidentalidad_2021.loc[:, 'Dia'] = accidentalidad_2021['Dia'].astype(int)

# Columna "Mes"
accidentalidad_2021.loc[:, 'Mes'] = accidentalidad_2021['Mes'].astype(int)

# Columna "Año"
accidentalidad_2021.loc[:, 'Ano'] = accidentalidad_2021['Ano'].astype(int)

# Columna "Choque"
accidentalidad_2021.loc[:, "Choque"] = accidentalidad_2021["Choque"].replace('X', 1.0)

# Columna "Atropello"
accidentalidad_2021.loc[:, "Atropello"] = accidentalidad_2021["Atropello"].replace('X', 1.0)

# Columna "Volcamiento"
accidentalidad_2021.loc[:, "Volcamiento"] = accidentalidad_2021["Volcamiento"].replace('X', 1.0)

# Columna "Caida de ocupante"
accidentalidad_2021.loc[:, "Caida de ocupante"] = accidentalidad_2021["Caida de ocupante"].replace('X', 1.0)

# Columna "Otro"
accidentalidad_2021.loc[:, "Otro"] = accidentalidad_2021["Otro"].replace('X', 1.0)

# Columna "Rural"
accidentalidad_2021.loc[:, "Rural"] = accidentalidad_2021["Rural"].replace('X', 1.0)

# Columna "Urbana"
accidentalidad_2021.loc[:, "Urbana"] = accidentalidad_2021["Urbana"].replace('X', 1.0)

# Columna "Particular"
accidentalidad_2021.loc[:, "Particular"] = accidentalidad_2021["Urbana"].replace('X', 1.0)

# Columna "Diplomatico"
accidentalidad_2021.loc[:, "Diplomatico"] = accidentalidad_2021["Diplomatico"].replace('X', 1.0)
```

```python
# Crear columna de edades  de involucrados
cols_edad = ["-18", "18-30", "31-60", "60"]

def extraer_edades(row):
    valores = []

    for col in cols_edad:
        val = str(row[col]).strip()

        if val in ["", "nan", "None"]:
            continue

        if "-" in val:               
            a, b = val.split("-")
            valores.append(a)
            valores.append(b)
        else:
            valores.append(val)      
    
    valores = [int(v) for v in valores if v.isdigit()]

    # aplicar reglas
    if len(valores) == 0:
        return pd.Series([0, 0])
    elif len(valores) == 1:
        return pd.Series([valores[0], 0])
    else:
        return pd.Series(valores[:2])   

# aplicar al dataframe
accidentalidad_2021[["Edad_inv_1", "Edad_inv_2"]] = accidentalidad_2021.apply(extraer_edades, axis=1)
```
